```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
def triangular_signal(amplitude, time):
    n = len(time)
    signal_array = np.zeros_like(time)
    signal_array[n//4:n//2] = amplitude*((2*time[n//4:n//2]/np.pi)+1)
    signal_array[n//2:3*n//4] = amplitude*((-2*time[n//2:3*n//4]/np.pi)+1)
    return signal_array

    
```

```python
time = np.linspace(-np.pi, np.pi, 1000)
```

```python
time = time.reshape(-1, 1)
```

```python
triangular_signal = triangular_signal(4, time)
```
